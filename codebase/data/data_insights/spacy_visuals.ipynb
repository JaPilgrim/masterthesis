```python
import spacy
from spacy import displacy
from IPython.core.display import display, HTML


def visualize_ner(text):
    nlp = spacy.load("de_core_news_sm")
    doc = nlp(text)
    display(HTML(displacy.render(doc, style="ent")))


# Example usage:
visualize_ner("Apple schaut sich den Kauf eines britischen Startups für 1 Milliarde Dollar an.")
```

```python
import spacy


def replace_entities_with_labels(text):
    nlp = spacy.load('de_core_news_lg')  # Or change to 'de_core_news_sm' for German
    doc = nlp(text)

    new_text = text
    for ent in doc.ents:
        new_text = new_text.replace(ent.text, ent.label_)

    return new_text


# Test the function
text = "Apple Inc. is planning to open a new store in San Francisco."
new_text = replace_entities_with_labels(text)
print(new_text)
```

```python
import spacy
import coreferee


def visualize_coreferences(text):
    nlp = spacy.load("de_core_news_sm")
    nlp.add_pipe('coreferee')
    doc = nlp(text)

    for cluster in doc._.coref_chains.clusters:
        print(f'Coreference: {cluster.main}, Referents: {cluster.mentions}')


# Example usage:
visualize_coreferences("Anna ging in den Park. Sie genoss das schöne Wetter.")
```
