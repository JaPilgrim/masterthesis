```python
import spacy
import de_core_news_sm

nlp = de_core_news_sm.load()
doc = nlp(processed_text)
    
def replace_corefs(doc):
    """
    Given a spaCy document, replaces every reference in a reference chain
    with the first occurrence in that chain.
    """
    # Iterate over every sentence in the document
    for sent in doc.sents:
        # Iterate over every token in the sentence
        for token in sent:
            # If the token is a pronoun or a named entity, and it has a coreference link
            if (token.pos_ == 'PRON' or token.ent_type_ != 0) and token._.in_coref:
                # Replace the token with the first mention in the chain
                chain = token._.coref_clusters[0]
                first_mention = chain.main
                token.sent.merge(first_mention.start, first_mention.end, first_mention.text, first_mention.ent_type_)
    return doc

doc = replace_corefs(doc)
print(doc.text)
```
