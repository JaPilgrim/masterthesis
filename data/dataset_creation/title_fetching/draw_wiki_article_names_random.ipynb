# Results: Random draw, thus not reproducible

```python


def get_random_german_wikipedia_title(i):
    url = "https://de.wikipedia.org/w/api.php"
    params = {"format": "json", "action": "query", "list": "random", "rnnamespace": 0, "rnlimit": 1}
    response = requests.get(url, params=params).json()
    title = response['query']['random'][0]['title']
    if i % 500 == 0:
        print(i)
    return title


max_len = 50000
with ThreadPoolExecutor(max_workers=10) as executor:
    random_list = list(executor.map(get_random_german_wikipedia_title, range(max_len)))

print(random_list)
```
