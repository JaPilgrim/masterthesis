```python


nlp = spacy.load("de_core_news_sm")
neuralcoref.add_to_pipe(nlp)


# Define a function that replaces all elements of a coreference chain with the first element
def replace_corefs(doc):
    for cluster in doc._.coref_clusters:
        # Get the first element of the coreference chain
        representative = cluster.main
        # Replace all other mentions of the same entity with the representative
        for mention in cluster.mentions:
            if mention != representative:
                mention[0].text = representative[0].text
    return doc


# Apply the `replace_corefs` function to a text
text = "Maria sah ihren Nachbarn. Er winkte ihr zu."
doc = nlp(text)
doc = replace_corefs(doc)
print(doc.text)
```

```python

nlp = de_core_news_sm.load()
doc = nlp(all_article_list_full.cleaned_article_text[0])


def replace_corefs(doc):
    """
    Given a spaCy document, replaces every reference in a reference chain
    with the first occurrence in that chain.
    """
    # Iterate over every sentence in the document
    for sent in doc.sents:
        # Iterate over every token in the sentence
        for token in sent:
            # If the token is a pronoun or a named entity, and it has a coreference link
            if (token.pos_ == 'PRON' or token.ent_type_ != 0) and token._.in_coref:
                # Replace the token with the first mention in the chain
                chain = token._.coref_clusters[0]
                first_mention = chain.main
                token.sent.merge(first_mention.start, first_mention.end, first_mention.text,
                                 first_mention.ent_type_)
    return doc


doc = replace_corefs(doc)
print(doc.text)
```
